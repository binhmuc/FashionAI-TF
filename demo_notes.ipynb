```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('outputs/yrd0528_test_b_stacked.csv')
```

```python
d = {'blouse': [0, 1, 2, 3, 4, 5, 6, 9, 10, 11, 12, 13, 14],
     'skirt': [15, 16, 17, 18],
     'outwear': [0, 1, 3, 4, 5, 6, 7, 8, 9, 11, 12, 13, 14],
     'dress': [0, 1, 2, 3, 4, 5, 6, 9, 10, 11, 12, 17, 18],
     'trousers': [15, 16, 19, 20, 21, 22, 23]}


def demo(line):
    cat = line['image_category']
    img = cv2.imread('test_b/' + line['image_id'])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    keypoints = d[cat]
    for i in keypoints:
        x, y, v = line[i+2].split('_')
        plt.scatter(int(x), int(y), marker='x', c=np.random.rand(3, 1))
    plt.imshow(img)
    plt.savefig('demo/demo_{}.jpg'.format(cat))
```

```python
demo(data.iloc[8516])
```
